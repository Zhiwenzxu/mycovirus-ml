1. Scans DNA sequence for all possible KMERs of length n

```python
from itertools import product
import pandas as pd
import numpy as np
```

```python
def possibleKMERS(n):
    return [p for p in product(["A","G","T","C"], repeat = n)]
```

```python
possibleKMERS(2)
```

```python
import readFasta
virus1 = readFasta.readFasta("data/Sclerotinia_biocontrol_mycovirus_nucleotide.fasta")
```

```python
def kmerCounter(test,n):
    pk = possibleKMERS(n)

    zeros = np.zeros((len(test),len(pk)))

    for row in range(0, len(test)):
        for column in range(1, len(test.iloc[0])-n+1):
            zeroIndex = 0
            for kmer in pk:
                if (kmer == tuple(test.iloc[row].iloc[column:column+n])):
                    zeros[row][zeroIndex] += 1
                    break
                zeroIndex += 1
            
    return pd.DataFrame(columns=pk, index=test.index, data = zeros)
```

```python
df1 = kmerCounter(pd.DataFrame([virus1.iloc[0]]),2)
```

```python
df2 = kmerCounter(pd.DataFrame([virus1.iloc[1]]),2)
```
